```python
uniqcosi = cosi[cosi['complex'] != cosi['simple']]
len(uniqcosi)
```

```python
uniqcompl = list(uniqcosi['complex'].values)
```

```python
uniqsimpl = list(uniqcosi['simple'].values)
```

```python
def tokenize_corpus(corpus):
	tokenized_corpus = []
	for sentence in corpus:
		tokenized_corpus.append(word_tokenize(sentence))
	return tokenized_corpus
```

```python
tokenized1 = tokenize_corpus(uniqcompl)
tokenized2 = tokenize_corpus(uniqsimpl)
```

```python
tokenized1[:10]
```

```python
tuniqcompl=[[word for word in sentence if word not in string.punctuation] for sentence in tokenized1]
```

```python
tuniqsimpl=[[word for word in sentence if word not in string.punctuation] for sentence in tokenized2]
```

```python
ziptuniq=(zip(tuniqcompl, tuniqsimpl))
```

```python
def del_larger_simples(complex_sentences, simple_sentences, threshold_for_no_of_words):
    lcomp=[]
    lsimp=[]
    for comp, simp in zip(complex_sentences, simple_sentences):
        if len(simp)-len(comp) <= threshold_for_no_of_words:
            lcomp.append(comp)
            lsimp.append(simp)
    return lsimp, lcomp
```

```python
compl_without_ls, simpl_without_ls = del_larger_simples(tuniqcompl, tuniqsimpl, 4)
```
